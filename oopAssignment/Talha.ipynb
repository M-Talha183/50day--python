```python
class Circle:
    def __init__(self, radius, color):
        """Initialize the circle with a radius and a color."""
        self.radius = radius
        self.color = color

    def setRadius(self, r):
        """Set the radius of the circle."""
        self.radius = r

    def getRadius(self):
        """Get the radius of the circle."""
        return self.radius

    def setColor(self, c):
        """Set the color of the circle."""
        self.color = c

    def getColor(self):
        """Get the color of the circle."""
        return self.color

    def getCircumference(self):
        """Calculate and return the circumference of the circle."""
        return 2 * 3.14159 * self.radius

    def getArea(self):
        """Calculate and return the area of the circle."""
        return 3.14159 * self.radius ** 2


# Create an instance of the Circle class with user-defined values
circle = Circle(4.0, "green")

# Access methods and display results
print("Radius:", circle.getRadius())
print("Color:", circle.getColor())
print("Circumference:", circle.getCircumference())
print("Area:", circle.getArea())

# Update the attributes using setter methods
circle.setRadius(7.0)
circle.setColor("yellow")

# Display updated results
print("\nUpdated Radius:", circle.getRadius())
print("Updated Color:", circle.getColor())
print("Updated Circumference:", circle.getCircumference())
print("Updated Area:", circle.getArea())
```

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        """Initialize the bank account with an initial balance."""
        self.balance_amount = initial_balance

    def deposit(self, amount):
        """Deposit an amount to the account."""
        if amount > 0:
            self.balance_amount += amount
            print(f"Deposited: {amount}. New balance: {self.balance_amount}")
        else:
            print("Deposit amount must be greater than 0.")

    def withdraw(self, amount):
        """Withdraw an amount from the account."""
        if 0 < amount <= self.balance_amount:
            self.balance_amount -= amount
            print(f"Withdrew: {amount}. New balance: {self.balance_amount}")
        elif amount > self.balance_amount:
            print("Insufficient funds.")
        else:
            print("Withdrawal amount must be greater than 0.")

    def balance(self):
        """Return the current balance."""
        return self.balance_amount


# Create an instance of BankAccount and test its methods
account = BankAccount(500)  # Initialize with a balance of 500

# Display the initial balance
print("Initial Balance:", account.balance())

# Deposit some money
account.deposit(200)

# Withdraw some money
account.withdraw(100)

# Attempt to withdraw more than the current balance
account.withdraw(700)

# Check the final balance
print("Final Balance:", account.balance())
```

```python
class Worker:
    def __init__(self, hoursWorked=0, wageRate=10):
        """Initialize the worker with default or user-defined values for hours worked and wage rate."""
        self.hoursWorked = hoursWorked  # Public attribute
        self.wageRate = wageRate        # Public attribute

    def setHoursWorked(self, h):
        """Set the hours worked by the worker."""
        if h >= 0:
            self.hoursWorked = h
            print(f"Hours worked set to: {self.hoursWorked}")
        else:
            print("Hours worked cannot be negative.")

    def changeRate(self, r):
        """Change the worker's hourly pay rate."""
        if r > 0:
            self.wageRate = r
            print(f"Wage rate changed to: {self.wageRate}")
        else:
            print("Hourly rate must be greater than 0.")

    def pay(self):
        """Calculate and return the pay as the product of hours worked and wage rate."""
        return self.hoursWorked * self.wageRate


# Test the Worker class
worker = Worker()  # Create a Worker object with default values

# Display default pay
print(f"Default pay: {worker.pay()}")

# Set hours worked
worker.setHoursWorked(40)

# Change hourly rate
worker.changeRate(15)

# Calculate and display pay
print(f"Total pay: {worker.pay()}")

# Update hours worked and rate, then calculate new pay
worker.setHoursWorked(45)
worker.changeRate(20)
print(f"Updated total pay: {worker.pay()}")
```

```python
class Int:
    def __init__(self, value=0):
        """Initialize the Int object with a default value of 0 or a user-defined value."""
        self.value = value  # The integer value for the class

    def set_value(self, value):
        """Set a new value for the Int object."""
        self.value = value

    def get_value(self):
        """Return the value of the Int object."""
        return self.value

    def add(self, other):
        """Add the value of another Int object to this one."""
        if isinstance(other, Int):
            return Int(self.value + other.get_value())
        else:
            print("Invalid type! The other object must be of type Int.")
            return None

    def display(self):
        """Display the current value of the Int object."""
        print(f"Value: {self.value}")


# Test the Int class
# Create an uninitialized Int object
int1 = Int()  # Default value is 0

# Create two initialized Int objects
int2 = Int(10)
int3 = Int(20)

# Display initial values
print("Initial Values:")
int1.display()
int2.display()
int3.display()

# Add int2 and int3, and store the result in int1
int1 = int2.add(int3)

# Display the result of the addition
print("\nAfter Addition:")
int1.display()
```

```python
class TollBooth:
    def __init__(self):
        self.total_cars = 0
        self.total_cash = 0

    def payingCar(self):
        self.total_cars += 1
        self.total_cash += 50

    def nopayCar(self):
        self.total_cars += 1

    def display(self):
        print(f"Total Cars: {self.total_cars}")
        print(f"Total Cash Collected: Rs. {self.total_cash}")


# Program to test the TollBooth class
def main():
    tollbooth = TollBooth()
    print("Tollbooth System")
    print("Enter 'p' for paying car, 'n' for non-paying car, any other key to display totals and exit.")

    while True:
        user_input = input("\nEnter your choice: ").lower()
        if user_input == 'p':
            tollbooth.payingCar()
            print("Paying car counted.")
        elif user_input == 'n':
            tollbooth.nopayCar()
            print("Non-paying car counted.")
        else:
            print("\nFinal Totals:")
            tollbooth.display()
            break


if __name__ == "__main__":
    main()
```

```python
class Time:
    def __init__(self, hours=0, minutes=0, seconds=0):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
        self._normalize_time()

    def _normalize_time(self):
        # Normalize seconds to minutes
        self.minutes += self.seconds // 60
        self.seconds = self.seconds % 60

        # Normalize minutes to hours
        self.hours += self.minutes // 60
        self.minutes = self.minutes % 60

        # Ensure hours are within a 24-hour range (optional for clock behavior)
        self.hours = self.hours % 24

    def display(self):
        print(f"{self.hours:02}:{self.minutes:02}:{self.seconds:02}")

    def addTime(self, other_time):
        self.hours += other_time.hours
        self.minutes += other_time.minutes
        self.seconds += other_time.seconds
        self._normalize_time()


# Instantiate and test the Time class
if __name__ == "__main__":
    t1 = Time(10, 59, 50)  # Arbitrary values
    t2 = Time(2, 30, 15)   # Arbitrary values

    print("Time t1:")
    t1.display()

    print("Time t2:")
    t2.display()

    # Add t2 to t1
    t1.addTime(t2)

    print("Time after adding t2 to t1:")
    t1.display()
```

```python
class Angle:
    def __init__(self, degrees=0, minutes=0.0, direction='N'):
        self.degrees = degrees
        self.minutes = minutes
        self.direction = direction

    def get_angle(self):
        """Obtain an angle value and direction from the user."""
        self.degrees = int(input("Enter degrees (0-180 for longitude, 0-90 for latitude): "))
        self.minutes = float(input("Enter minutes (0.0-59.9): "))
        self.direction = input("Enter direction (N, S, E, W): ").strip().upper()

        # Validate inputs
        if not (0 <= self.degrees <= 180) or not (0.0 <= self.minutes < 60):
            print("Invalid degrees or minutes. Please try again.")
            self.get_angle()

        if self.direction not in ('N', 'S', 'E', 'W'):
            print("Invalid direction. Please enter N, S, E, or W.")
            self.get_angle()

    def display(self):
        """Display the angle in 179°59.9’ E format."""
        print(f"{self.degrees}\u00B0{self.minutes:.1f}’ {self.direction}")


# Main program to test the Angle class
if __name__ == "__main__":
    # Display an angle initialized with the constructor
    initial_angle = Angle(149, 34.8, 'W')
    print("Initialized Angle:")
    initial_angle.display()

    # Loop to allow user input and display
    while True:
        print("\nEnter a new angle (or type 'exit' to quit):")
        user_input = input("Type 'exit' to quit or press Enter to continue: ").strip().lower()
        if user_input == 'exit':
            break

        user_angle = Angle()
        user_angle.get_angle()
        print("Entered Angle:")
        user_angle.display()
```

```python
class Angle:
    def __init__(self, degrees=0, minutes=0.0, direction='N'):
        self.degrees = degrees
        self.minutes = minutes
        self.direction = direction

    def get_angle(self, angle_type="latitude"):
        """Obtain an angle value and direction from the user."""
        if angle_type == "latitude":
            max_degrees = 90
            directions = ('N', 'S')
        else:  # longitude
            max_degrees = 180
            directions = ('E', 'W')

        try:
            self.degrees = int(input(f"Enter degrees (0-{max_degrees}) for {angle_type}: "))
            self.minutes = float(input("Enter minutes (0.0-59.9): "))
            self.direction = input(f"Enter direction ({'/'.join(directions)}): ").strip().upper()
        except ValueError:
            print("Invalid input. Please enter numeric values for degrees and minutes.")
            self.get_angle(angle_type)

        # Validate inputs
        if not (0 <= self.degrees <= max_degrees and 0.0 <= self.minutes < 60):
            print("Invalid degrees or minutes. Please try again.")
            self.get_angle(angle_type)

        if self.direction not in directions:
            print(f"Invalid direction. Please enter {directions[0]} or {directions[1]}.")
            self.get_angle(angle_type)

    def display(self):
        """Display the angle in 179°59.9’ E format."""
        return f"{self.degrees}\u00B0{self.minutes:.1f}’ {self.direction}"


class Ship:
    count = 0  # Class attribute to keep track of ship numbers

    def __init__(self):
        Ship.count += 1
        self.ship_number = Ship.count
        self.latitude = Angle()
        self.longitude = Angle()

    def set_position(self):
        """Get the ship's position (latitude and longitude) from the user."""
        print(f"Setting position for Ship {self.ship_number}:")
        self.latitude.get_angle("latitude")
        self.longitude.get_angle("longitude")

    def report_position(self):
        """Report the ship's number and position."""
        print(f"Ship {self.ship_number} is at:")
        print(f"  Latitude: {self.latitude.display()}")
        print(f"  Longitude: {self.longitude.display()}")


# Main program
if __name__ == "__main__":
    # Create three Ship objects
    ship1 = Ship()
    ship2 = Ship()
    ship3 = Ship()

    # Input positions for each ship
    ship1.set_position()
    ship2.set_position()
    ship3.set_position()

    # Display positions for each ship
    print("\nShip Positions:")
    ship1.report_position()
    ship2.report_position()
    ship3.report_position()
```
